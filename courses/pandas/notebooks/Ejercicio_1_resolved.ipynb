### EJERCICIO 6: Top 10 de gestores, para un conjunto del segmento global, que tengan una mayor evolución en saldo en fondos con cod_seglobal= 41
Evolucion: Porcentaje de aumento en el importe gestionado entre Julio y Diciembre del 2015

```python
# noTA ESTUDIAR SI ES MEJOR EL GROUP BY MAS RESET INDEX O NO
A =fondos_ev.groupby(['COD_SEGLOBAL','COD_USUARIO']).sum()
#A=fondos_ev
A["EVOLUCION"]=(A.fondos_201512-A.fondos_201507)/A.fondos_201507
A_41 =A.query("COD_SEGLOBAL==41.0")
top_ten = A_41.sort_values(by='EVOLUCION',ascending=False)
top_ten.reset_index()[["COD_USUARIO","EVOLUCION"]].head(10)
#top_ten[["COD_USUARIO","EVOLUCION"]].head(10)
```

### EJERCICIO 8 : Crear una tabla donde a nivel usuario se muestre la evolucion total en fondos por meses, la cantidad total en fondos del 2015, el número de aportaciones realizadas por sus clientes, y la cantidad total de las aportaciones. A continuación sacar el top 3 de los usarios con mayor número de aportaciones y mayor cantidad en fondos


```python
fondos_tablon = fondos_ev.groupby(['COD_USUARIO']).sum().reset_index()
del fondos_tablon['COD_SEGLOBAL']
fondos_tablon['TOTAL_fondos'] = fondos_tablon.sum(axis=1)
#df2 = extr_fondos[['COD_USUARIO','imp_trans']].groupby(['COD_USUARIO']).size().reset_index(name='N_aportaciones')
df2 = extr_fondos[['COD_USUARIO','imp_trans']].groupby('COD_USUARIO')['imp_trans'].agg(['sum','count']).reset_index()
tablon = fondos_tablon.merge(df2, how='outer', on=['COD_USUARIO']).fillna(value = 0)
tablon
```

```python
tablon.sort_values(by=['TOTAL_fondos','sum']).tail(3)
```
